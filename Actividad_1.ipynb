```python
import numpy as np

def generar_saldo_inicial():
    return np.random.randint(1000, 10000)

saldo = generar_saldo_inicial()
print("Bienvenido al cajero automatico")
print("\nRegistrarse:")
Docu= int(input("\nIngrese su documento: "))
ContraseñaU = int(input("Ingrese su contraseña: "))
print("\nSe ha registrado exitosamente")

def consultar_saldo():
    print(f"Su saldo actual es: ${saldo:.2f}")
def depositar():
    global saldo
    monto = float(input("Ingrese el monto a depositar: "))
    if monto <= 0:
        print("El monto debe ser mayor que cero.")
        return
    else:
        saldo += monto
        print(f"Ha depositado ${monto:.2f}. Su nuevo saldo es: ${saldo:.2f}")

def retirar():
    global saldo
    monto = float(input("Ingrese el monto a retirar: "))
    if monto <= 0:
        print("El monto debe ser mayor que cero.")
        return
    else:
      if monto > saldo:
        print("Saldo insuficiente.")
      else:
        saldo -= monto
        print(f"Ha retirado ${monto:.2f}. Su nuevo saldo es: ${saldo:.2f}")

def transferir():
    global saldo
    global saldo_otro_usuario
    monto = float(input("Ingrese el monto a transferir: "))
    if monto <= 0:
        print("El monto debe ser mayor que cero.")
        return
    else:
      if monto > saldo:
        print("Saldo insuficiente.")
      else:
        saldo -= monto
        saldo_otro_usuario += monto
        print(f"Ha transferido ${monto:.2f}. Su nuevo saldo es: ${saldo:.2f}")

def ingresar_contraseña():
    intentos = 3
    while intentos > 0:
        entrada_doc = int(input("\nIngrese su documento: "))
        entrada_contraseña = int(input("Ingrese su contraseña: "))

        if entrada_doc == Docu and entrada_contraseña == ContraseñaU:
            print("\nUsuario autenticado correctamente.")
            return True
        else:
            intentos -= 1
            print(f"Credenciales incorrectas. Intentos restantes: {intentos}")

    print("Acceso denegado.")
    return False


def cajero_automatico():
    if not ingresar_contraseña():
        return

    while True:
        print("\nBienvenido al cajero automático")
        print("1. Consultar saldo")
        print("2. Depositar dinero")
        print("3. Retirar dinero")
        print("4. Transferir dinero")
        print("5. Salir")

        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            consultar_saldo()
        elif opcion == "2":
            depositar()
        elif opcion == "3":
            retirar()
        elif opcion == "4":
            transferir()
        elif opcion == "5":
            print("\nGracias por usar el cajero automático. Adiós.")
            break
        else:
            print("Opción no válida. Intente de nuevo.")

cajero_automatico()
```
